```python
#Q1 fbonacci
def fibo(n, memo):
    if n==0:
        return 0
    elif n ==1:
        return 1
    elif n<0:
        return "-ve number"
    
    if n not in memo:
            memo[n] =  fibo(n-1, memo) + fibo(n-2, memo)
            
    return memo[n]


memo={}
print("-->",fibo(50,memo))
memo
```

```python
def grid(m,n, memo):
    if m==0 or n==0:
        return 0
    if m==1 and n==1:
        return 1
    
    if (m,n) not in memo:
        memo[m,n] = grid(m-1,n,memo) + grid(m,n-1,memo)
        
    return memo[m,n]
    
memo = {}   
grid(18,18,memo)
memo
```

```python
def HowSum(targetvalue, arr, memo):
    if targetvalue in memo:
        return memo[targetvalue]
    if targetvalue == 0 :
        return []

    if targetvalue < 0:
        return None
    
    for i in range(len(arr)):
        
        temp = targetvalue - arr[i]
        remai = HowSum(temp, arr, memo)
        if remai != None:
            memo[targetvalue] = [remai, arr[i]]
            return memo[targetvalue]
        
    memo[targetvalue] = None
    return None




memo = {}
print(HowSum(7, [2,3,5], memo))
```

```python
def bestSum(target, array, memo):
    if target in memo:
        return memo[target]
    if target == 0:
        return []
    if target < 0:
        return None
    
    leng = None
    
    for i in range(len(array)):
        diff = target - array[i]
        temp = bestSum(diff, array, memo)
        
        if temp != None:
            combo = [temp, array[i]]
            
            
            if leng == None or len(combo) <= len(leng) :
                leng = combo
    memo[target] = leng   
    
    return leng
        
    
        
        
        
        
memo = {} 
bestSum(100, [2,5,25], memo)    
memo
```
