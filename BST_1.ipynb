```python
#c 214   and C216        Search an element is tree and find the minimum element is the tree

class BST:
    def __init__(self, data):
        self.data = data
        self.leftchild = None
        self.rightchild = None
        
    
    def create_node(self, rootnode, nodevalue):
        self.rootnode= rootnode
        if rootnode == None:
            rootnode = nodevalue
            
        if nodevalue < rootnode.data:
            if rootnode.leftchild == None:
                rootnode.leftchild = BST(nodevalue)
                
            else:
                self.create_node(rootnode.leftchild, nodevalue )
                
        if nodevalue > rootnode.data:
            if rootnode.rightchild == None:
                rootnode.rightchild = BST(nodevalue)
                
            else:
                self.create_node(rootnode.rightchild, nodevalue)
                
    
                    
                
    def pre_traverse(self, rootnode):
        self.rootnode = rootnode
        if rootnode == None:
            return None
        print(rootnode.data)
        self.pre_traverse(rootnode.leftchild)
        self.pre_traverse(rootnode.rightchild)
        
        
    
    
    def search_key(self, rootnode, key):
        self.rootnode = rootnode
        if rootnode == None:
            return False
        
        if key == rootnode.data:
            return True
        
        if key< rootnode.data:
            return self.search_key(rootnode.leftchild, key)
            
        if key > rootnode.data:
            return self.search_key(rootnode.rightchild, key)
            
        else:
            return False
        
        
    def minimum_key(self, rootnode):
        self.rootnode = rootnode
        
        current = rootnode
        
        while current.leftchild != None:
            current = current.leftchild
            
        return current.data

        
    def maximum_key(self, rootnode):
        self.rootnode = rootnode
        
        current = rootnode
        
        while current.rightchild != None:
            current = current.rightchild
            
        return current.data
        
        
        
     
a = BST(5)
a.create_node(a, 9)
a.create_node(a, 3)
a.create_node(a, 4)
a.create_node(a, 2)
a.create_node(a, 7)
root = a
#a.pre_traverse(root)

print(a.minimum_key(root))
print(a.maximum_key(root))
```
