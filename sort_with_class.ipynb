```python
class sorting:
   
    def __init__(self,data):
        print("the given array is: " , data)
        self.arr = data
        
   
    def bubble_sort(self):
        print("This is bubble sort: ")
        i=1
        for f in range(len(self.arr)-1):
            for r in range (len(self.arr)-1 -f):
                if self.arr[r] > self.arr[r+1]:
                    self.arr[r] , self.arr[r+1] = self.arr[r+1], self.arr[r]
                    print(self.arr)
                    print("swap done!")
                    i = i+1
                    print(i)
                else:
                
                    print(self.arr)
                    print("no swap! ")
        print(i)
        return self.arr  
    
    def selection_sort(self):
        print("This is selection sort: ")
        for j in range(len(self.arr)):
            min_value = j
            for z in range(j+1 , len(self.arr)):
                if self.arr[min_value] > self.arr[z]:
                    min_value = z
            self.arr[j] , self.arr[min_value] =self.arr[min_value] , self.arr[j]   
        print(self.arr)
    
    def insertion_sort(self):
        print("this is insertion sort: ")
        for i in range(1 , len(self.arr)):
            position = i
            currentvalue = self.arr[i]
        
            while position > 0 and self.arr[position - 1] > currentvalue:
                self.arr[position] = self.arr[position - 1]
                position = position-1
            
            self.arr[position] = currentvalue
        return(self.arr)  
    
```
