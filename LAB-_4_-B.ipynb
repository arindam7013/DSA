# Gradient Descent Implementation

```python
def gradient_descent(x,y):
    slope = intercept = 0
    iterations = 1000
    n = len(x)
    learning_rate = 0.001

    for i in range(iterations):
        y_predicted = slope * x + intercept
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        m = -(2/n)*sum(x*(y-y_predicted))
        c = -(2/n)*sum(y-y_predicted)
        slope = slope - learning_rate * m
        intercept = intercept - learning_rate * c
        print ("iteration {}: slope--> {}, intercept--> {}, cost--> {}".format(i+1, slope, intercept, cost))
        

x = np.array(df.Present_Price[:50])
y = np.array(df.Selling_Price[:50])

gradient_descent(x,y)
```

```python
import matplotlib.pyplot as plt

x = df.Present_Price[:50].values 
slope = 0.6781894045399225
intercept = -0.08518158878086576
y_pred = []

for i in x:
    y_pred.append(slope * i + intercept)

plt.scatter(x,y)
plt.plot(x, y_pred, color='red')
plt.show()
```
