```python
#K Levels down , node to root , Print Nodes K Level Far
class BinaryTree:
    
    
    
    def __init__(self, data):
        self.data = data
        self.leftchild = None
        self.rightchild = None
        self.path = []
        
        
    def preOrder(self,root):
        if not root:
            return
    
        print(root.data)
        self.preOrder(root.leftchild)
        self.preOrder(root.rightchild)
    
    
    
                
            
    
    
    
    
    
    def postOrder(self, root):
        if not root:
            return
        
        self.postOrder(root.leftchild)
        self.postOrder(root.rightchild)
        print(root.data)
    
    
    
    def inOrder(self, root):
        if not root:
            return
    
    
        self.inOrder(root.leftchild)
        print(root.data)
        self.inOrder(root.rightchild)
        
        
    def search(self, inorder, start, end, curr):
        
        for i in range(start , end+1):
            if inorder[i] == curr:
                return i
            
        return -1 
        
        
    
    def countLeaves(self, root):
        if root == None:
            return 0
        if root.leftchild == None and root.rightchild == None:
            return 1
        
        l = self.countLeaves(root.leftchild)
        r = self.countLeaves(root.rightchild)
        
        return l + r
    
    
    
    
    def level_k(self , root , k):
        if root == None:
            return
        
        if k == 0:
            print(root.data)
        
        self.level_k(root.leftchild , k-1)
        self.level_k(root.rightchild , k-1)
                
     
    
    
    
    
    def k_level(self, root, k, blocker):
        
        if root == None or k < 0 or root == blocker:
            return
        
        if k == 0:
            print(root.data)
        
        
        self.k_level(root.leftchild, k-1, blocker)
        self.k_level(root.rightchild , k-1, blocker)
        
        
        
    def find_node_path_final(self, root, key):
        if root == None:
            return False
        
        if root.data == key:
            self.path.append(root.data)
            
            return self.path
        
        left = self.find_node_path(root.leftchild, key)
        if (left):
            self.path.append(root.data)
            
            return   self.path
        
        right = self.find_node_path(root.rightchild, key)
        if (right):
            self.path.append(root.data)
            
            return  self.path
        
        
        else:
            return False
        
        
    def find_node_path(self, root, key , path = []):
        if root == None:
            return False
        
        if root.data == key:
            path.append(root.data)
            
            return  path
        
        left = self.find_node_path(root.leftchild, key)
        if (left):
            path.append(root.data)
            
            return  path
        
        right = self.find_node_path(root.rightchild, key)
        if (right):
            path.append(root.data)
            
            return  path
        
        
        else:
            return False 
        
        
    def all_nodes_k(self, root, node, k, blocker):
        j = self.find_node_path(root , node)
        
        for i in range(len(j)):
            if i == 0:
                blocker = None
                
                self.k_level(j[i], k-i, blocker)
                
            else:
                self.k_level(j[i], k-i, j[i-1])
        
        

    
    
N1 = BinaryTree("N1")
N2 = BinaryTree("N2")
N3 = BinaryTree("N3")
N4 = BinaryTree("N4")
N5 = BinaryTree("N5")
N6 = BinaryTree("N6")
N7 = BinaryTree("N7")
N8 = BinaryTree("N8")
N9 = BinaryTree("N9")
N10 = BinaryTree("N10")
N11 = BinaryTree("N11")



N1.leftchild = N2
N1.rightchild = N3
N2.leftchild = N4
N2.rightchild = N5
N3.leftchild = N6
N3.rightchild = N7
N4.leftchild = N8
N4.rightchild = N9
N5.leftchild = N10
N5.rightchild = N11

rootnode = N1


#k = N1.find_node_path(rootnode , 'N8')
print("-->", k)
N1.level_k(N1 , 3)
```

```python

class Solution:
    
    
    def isBalanced(self, root: Optional[TreeNode]) -> bool:
        
        
        
        def ht(root):
            if root == None:
                return 0

            l = ht(root.left)
            r = ht(root.right)
            gap = abs(l - r)
            if l < 0 or r < 0:
                return -1
            if gap > 1:
                return -1
                
            return 1 + max(l , r) 
            
            
        x = ht(root)
        if x < 0:
            return False
        else:
            return True
```

```python
#Size, Sum, Max and Height in a Binary Tree
class BinaryTree:
    
    
    
    def __init__(self, data):
        self.data = data
        self.leftchild = None
        self.rightchild = None
        
        
    def preOrder(self,root):
        if not root:
            return
    
        print(root.data)
        self.preOrder(root.leftchild)
        self.preOrder(root.rightchild)
    
    
    def summ(self, root):
        if root == None:
            return 0
        
        sl = self.summ(root.leftchild)
        sr = self.summ(root.rightchild)
        
        return sl + sr + root.data
    
    
    def maxi(self, root):
        if root == None:
            return 0
        
        sl = self.maxi(root.leftchild)
        sr = self.maxi(root.rightchild)
        
        return max(sl , sr , root.data)
    
    
    def height(self, root):
        if root == None:
            return 0
        
        sl = self.height(root.leftchild)
        sr = self.height(root.rightchild)
        
        return max(sl , sr) + 1   
    
    def size(self, root):
        if root == None:
            return 0
        
        sl = self.size(root.leftchild)
        sr = self.size(root.rightchild)
        
        return sl + sr + 1
    
    
    def postOrder(self, root):
        if not root:
            return
        
        self.postOrder(root.leftchild)
        self.postOrder(root.rightchild)
        print(root.data)
    
    
    
    def inOrder(self, root):
        if not root:
            return
    
    
        self.inOrder(root.leftchild)
        print(root.data)
        self.inOrder(root.rightchild)
        
        
    def search(self, inorder, start, end, curr):
        
        for i in range(start , end+1):
            if inorder[i] == curr:
                return i
            
        return -1 
        
        
    
    def countLeaves(self, root):
        if root == None:
            return 0
        if root.leftchild == None and root.rightchild == None:
            return 1
        
        l = self.countLeaves(root.leftchild)
        r = self.countLeaves(root.rightchild)
        
        return l + r
        
        

    
    
N1 = BinaryTree(1)
N2 = BinaryTree(2)
N3 = BinaryTree(33)


N1.leftchild = N2
N1.rightchild = N3


N4 = BinaryTree(4)
N5 = BinaryTree(5)


N2.leftchild = N4
N2.rightchild = N5

rootnode = N1  



N1.height(N1)
#N1.maxi(N1)
#N1.size(N1)
#N1.summ(N1)
#N1.countLeaves(N1)
```
